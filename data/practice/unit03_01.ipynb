```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
result = [[] for _ in range(3)]
for row in data:
    if row[2].find('05-30') != -1:
        if row[3] and row[4] and row[6]:
            result[0].append(float(row[6]))
            result[1].append(float(row[3]))
            result[2].append(float(row[4]))
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('5월 30일 기온')
plt.plot(result[0], 'b', label='최저')
plt.plot(result[1], 'g', label='평균')
plt.plot(result[2], 'r', label='최고')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
max_temp = [[] for _ in range(13)]
for row in data:
    if row[4]:
        max_temp[int(row[2].split('-')[1])].append(float(row[4]))
plt.figure(figsize=(35,25))
for i in range(1, 13):
    plt.hist(max_temp[i], bins=100, label=f'{i}')
plt.legend()

plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
jan = []
aug = []
for row in data:
    if row[4]:
        if row[2].split('-')[1] == '01':
            jan.append(float(row[4]))
        elif row[2].split('-')[1] == '08':
            aug.append(float(row[4]))
plt.boxplot(jan)
plt.boxplot(aug)
plt.show()
plt.boxplot([jan, aug])
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
result = [[] for _ in range(32)]
for row in data:
    if row[4]:
        if row[2].split('-')[1] == '08':
            result[int(row[2].split('-')[2])].append(float(row[4]))
plt.style.use('ggplot')
plt.figure(figsize=(25,15), dpi=300)
plt.boxplot(result[1:], showfliers=False)
plt.show()
```
