```python
import csv
import matplotlib.pyplot as plt


f = open('population.csv')
data = csv.reader(f)
header = next(data)
for row in data:
    if row[0].find('신월3동') != -1:
        population = list(map(int, row[3:]))
plt.plot(population, 'skyblue')
plt.show()
plt.figure(figsize=(20,10))
plt.bar(range(len(population)), population)
plt.show()
plt.barh(range(len(population)), population)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt


def minus(num):
    return -num


f = open('population_gender.csv')
data = csv.reader(f)
header = next(data)
for row in data:
    if row[0].find('신월3동') != -1:
        male = list(map(int, row[3:104]))
        female = list(map(int, row[-1:-102:-1][::-1]))        
plt.barh(range(len(female)), list(map(minus, female)), color='r')
plt.barh(range(len(male)), male, color='b')
plt.show()
```

```python
import matplotlib.pyplot as plt


plt.axis('equal')
data = [10, 20, 15, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
plt.pie(data, labels=labels, autopct='%.1f%%', colors=colors, explode=(0, 0, 0.1, 0), startangle=90)
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import math


f = open('population_gender.csv')
data = csv.reader(f)
header = next(data)
name = input()
for row in data:
    if name in row[0]:
        m = [int(i.replace(',','')) for i in row[3:104]]
        f = [int(i.replace(',','')) for i in row[-1:-102:-1][::-1]]
        total = list(map(sum, zip(m, f)))
        break
plt.style.use('ggplot')
plt.figure(figsize=(15,10))
plt.scatter(m, f, c=range(len(m)), cmap='jet', alpha=0.5)
plt.plot([0, max(max(m), max(f))], [0, max(max(m), max(f))], 'g:', alpha=0.5)
plt.colorbar()
plt.show()
plt.style.use('ggplot')
plt.figure(figsize=(15,10))
plt.scatter(m, f, c=range(len(m)), cmap='jet', alpha=0.5, s=list(map(math.sqrt, total)))
plt.plot([0, max(max(m), max(f))], [0, max(max(m), max(f))], 'g:', alpha=0.5)
plt.colorbar()
plt.show()
```
